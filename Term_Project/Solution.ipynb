# __One Electron 1D Harmonic Oscillator__

```python
from Library import *
import numpy as np
```

```python
n=100
T=[[0 for i in range(n)] for i in range(n)]
V=[[0 for i in range(n)] for i in range(n)]
H=[[0 for i in range(n)] for i in range(n)]
x_i=-5
x_f=5
xlist=[]
mu=1
k=1
delta = (x_f-x_i)/(n-1)
for i in range(len(T)):
    for j in range(len(T)):
            if i==j:
                V[i][j]=(0.5)*k*(x_i**2)
                T[i][j]= (np.pi**2)/(6*mu*(delta**2))
            else:
                V[i][j]=0
                T[i][j]=((-1)**(i-j))/(((i-j)**2)*mu*(delta**2))
    xlist.append(x_i)            
    x_i+=delta

for i in range(len(T)):
     for j in range(len(T)):
         H[i][j]=T[i][j]+V[i][j]  


eigen_values, eigen_vectors = np.linalg.eig(T)
import matplotlib.pyplot as plt

for i in range(len(eigen_vectors[:,0])):
    eigen_vectors[:,0][i]=eigen_vectors[:,0][i]**2
    eigen_vectors[:,1][i]=eigen_vectors[:,1][i]**2
plt.plot(xlist,eigen_vectors[:,0],label='Ground state')
plt.plot(xlist,eigen_vectors[:,1],label='First excited state')
plt.grid()
plt.legend()
plt.legend(loc='upper left')

plt.show() 
```
