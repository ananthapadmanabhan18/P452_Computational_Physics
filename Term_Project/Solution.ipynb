# __One Electron 1D Harmonic Oscillator__

```python
from Library import *
import numpy as np
```

Defining the constants

```python
h=6.626*10**(-34) #J*s
m=9.109*10**(-31) #kg
e=1.602*10**(-19) #C
h_cross=h/(2*3.1415926)
```

```python
W=np.pi*100
E=h_cross*W
a1=2*m*E/(h_cross**2)
a2=(m*m*W*W)/(h_cross**2)
```

Defining the functions

```python
def dshi(r,x):
    return r[1]
def dphi(r,x):
    return (a2*x**2-a1)*r[0]
flist=[dshi,dphi]
```

```python
y0s=[0,0]
a,b=gen_RK4_coupled(flist,0,y0s,10,0.01)
import matplotlib.pyplot as plt
plt.plot(a,b[0],"-b", label="y(t)")
plt.plot(a,b[1],"-r", label="y'(t)")
plt.title('Plot of y and dy/dt vs t')
plt.legend(loc="upper left")
plt.xlabel('time (t)')
plt.ylabel('position(y)/Velocity (v=dy/dt))')
plt.legend()
plt.show()
```
