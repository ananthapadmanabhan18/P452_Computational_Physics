# __One Electron 2D Harmonic Oscillator__

```python
from Library import *
```

Defining the constants

```python
h=6.626*10**(-34) #J*s
m=9.109*10**(-31) #kg
e=1.602*10**(-19) #C
h_cross=h/(2*3.1415926)
```

Defining the Energies and Frequency

```python
E1=10**(-3)*e #J
omega=10**12*2*3.1415926 #Hz
```

Defining:
$\frac{\partial X}{\partial x}=\phi (x)$
and
$\frac{\partial \phi}{\partial x}=(E_1 - \frac{m\omega^2 x^2}{2})(\frac{-2mX}{\hbar^2})$

```python
def dX(ilist,x):
    return ilist[1]

def dphi(ilist,x):
    return (E1-(0.5*m*omega*omega*x*x))*((-2*m*ilist[0])/(h_cross**2))

flist=[dX,dphi]
y0s=[0,1]
a,b=gen_RK4_coupled(flist,0,y0s,10,0.01)

import matplotlib.pyplot as plt
plt.plot(a,b[0],"-b", label="X")
plt.plot(a,b[1],"-r", label=r"$\phi$")
plt.title(r'Plot of X and d$\phi$/dt vs t')
plt.legend(loc="upper left")
plt.xlabel('x')
plt.ylabel(r'$\phi$')
plt.legend()
plt.show()
```
