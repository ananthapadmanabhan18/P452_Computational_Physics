```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')
# plt.style.use(['science','ieee'])
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(11,6))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple plot')
plt.show()
```
