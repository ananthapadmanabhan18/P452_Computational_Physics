# Harmonic Oscillator

```python
import numpy as np
'''
Grid Points
'''
n=200
x = np.linspace(-5,5,n)
h = x[1]-x[0] 

'''
Defining the 2nd Derivative Operator
'''
D=(-np.eye(n)+np.diagflat(np.ones(n-1),1))/h
D2=D.dot(-D.T)
D2[-1,-1]=D2[0,0]

'''
Defining the Harmonic Potential Matrix
'''
V_harm=np.diagflat(x*x/2)

'''
Finding the Eigenvalues and Eigenvectors
'''
Energy_harmonic,psi_harmonic = np.linalg.eigh(-D2/2+V_harm)  

'''
Plotting the Eigenvalues
'''
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','ieee'])
plt.figure(figsize=(8,5))
for i in range(3):
    plt.plot(x,psi_harmonic[:,i],label=f"E={Energy_harmonic[i]:.2f}")
plt.xlim(-5, 5)
# plt.grid()
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.legend()
plt.xlabel("Position (x)")
plt.ylabel(r"$\Psi(x)$")
plt.title("Solutions to the Harmonic Oscillator: $(-1/2)d^2/dx^2 +  x^2 /2 = 0$")
plt.show()  

del plt, np, h, x, D, D2, V_harm, Energy_harmonic, psi_harmonic, i, n
```

# Half Harmonic Oscillator

```python
import numpy as np
'''
Grid points
'''
n=600
x = np.linspace(-5,5,n)
h = x[1]-x[0] 


'''
Second derivative matrix
'''
D=(-np.eye(n)+np.diagflat(np.ones(n-1),1))/h
D2=D.dot(-D.T)
D2[-1,-1]=D2[0,0]


'''
Potential energy matrix
'''
def V_half(x):    # half harmonic oscillator potential
    if x<0:
        return 10000000000
    else:
        return 0.5*x**2

V_half = np.array([V_half(xi) for xi in x])
V_half = np.diagflat(V_half)
# print(V_half)
'''
Find the eigenvalues and eigenvectors
'''
Energy_half,psi_half = np.linalg.eigh(-D2/2 + V_half)


'''
Plot the results
'''
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
for i in range(3):
    plt.plot(x,psi_half[:,i],label=f"E={Energy_half[i]:.2f}")
plt.xlim(-5,5)
# plt.grid()
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.legend()
plt.xlabel("Position (x)")
plt.ylabel(r"$\Psi_{HH}(x)$")
plt.title("Solutions to the Half Harmonic Oscillator")
plt.show()  
del plt, np, h, x, D, D2, V_half, Energy_half, psi_half, i, n,
```

# Some hartree Like potential

```python
import numpy as np
'''
Grid points
'''
n=600
x = np.linspace(-5,5,n)
h = x[1]-x[0] 


'''
Second derivative matrix
'''
D=(-np.eye(n)+np.diagflat(np.ones(n-1),1))/h
D2=D.dot(-D.T)
D2[-1,-1]=D2[0,0]


'''
Potential energy matrix
'''
def V_1(x):    
    return 0.5*(x**2) + 1/np.sqrt(x*x+0.01)

V_half = np.array([V_1(xi) for xi in x])
V_half = np.diagflat(V_half)
# print(V_half)
'''
Find the eigenvalues and eigenvectors
'''
Energy_1,psi_1 = np.linalg.eigh(-D2/2 + V_half)


'''
Plot the results
'''
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
for i in range(3):
    plt.plot(x,psi_1[:,i],label=f"E={Energy_1[i]:.2f}")
plt.xlim(-5,5)
# plt.grid()
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.legend()
plt.xlabel("Position (x)")
plt.ylabel(r"$\Psi_{HH}(x)$")
plt.title(r"Solutions to $(\frac{-1}{2})D^2x + (\frac{1}{2})x^2 +\frac{1}{\sqrt{1+x^2}}=0$")
plt.show()  
```
